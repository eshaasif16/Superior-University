```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from PyPDF2 import PdfReader

reader = PdfReader("FHWA-NJ-2018-007.pdf")
# reader = PdfReader("Library-Policies-and-Procedures-Handbook91619.pdf")

text = "\n".join([page.extract_text() for page in reader.pages if page.extract_text()])

model = SentenceTransformer('all-MiniLM-L6-v2')

# spliting into chunks
chunks = text.split("\n\n")
embeddings = model.encode(chunks)

dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings))
```

```python
def ask_bot(query):
    query_embedding = model.encode([query])
    D, I = index.search(np.array(query_embedding), k=1)
    return chunks[I[0][0]]

print(ask_bot("What are the library opening hours?"))
```

```python
print(ask_bot("what are the policy?"))
```

```python
print(ask_bot("What is the library’s policy on late returns?"))
```

```python
print(ask_bot("What is the procedure for borrowing books?"))
```
